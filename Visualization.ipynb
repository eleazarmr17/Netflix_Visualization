# Netflix: Visualization

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_Netflix = pd.read_csv("Data/netflix_outer.csv")
df_Netflix.head()
```

We are not going to create a model or something like that, we only want to practice some visualization techniques so we will delete all the na

```python
sns.heatmap(df_Netflix.isnull(),yticklabels=False, cbar=False, cmap='viridis')
plt.show()
```

```python
df=df_Netflix.dropna()
df
```

```python
sns.heatmap(df.isnull(),yticklabels=False, cbar=False, cmap='viridis')
plt.show()
```

Now we dont have any missing value

```python
df.describe()
```

```python
df.info()
```

We have in our dataset without missings 208 registers

```python
sns.countplot(x="type",data=df_Netflix, palette='rocket')
plt.title('Quantity of titles (Movie vs Tv-Show)', fontsize=16)
plt.xlabel('Type', fontsize=12)
plt.ylabel('Quantity', fontsize=12)
plt.show()
```

```python
rating_order = df['rating'].value_counts().sort_values(ascending=False).index

sns.countplot(x="rating",data=df, palette='hls', order=rating_order)
plt.title('Rate distribution', fontsize=16)
plt.xlabel('', fontsize=12)
plt.ylabel('Quantity', fontsize=12)
plt.xticks(rotation=45) 
plt.show()
```

```python
country_order = df['country'].value_counts().sort_values(ascending=False).index
top_10_countries = country_order[:10]

sns.countplot(x="country",data=df, palette='hls', order=top_10_countries)
plt.title('Top 10 countries with more titles', fontsize=16)
plt.xlabel('', fontsize=12)
plt.ylabel('Quantity', fontsize=12)
plt.xticks(rotation=45) 
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x="rating", y="IMDB Score",data=df, palette='mako')
plt.title('Average IMDB Score', fontsize=16)
plt.xlabel('', fontsize=12)
plt.ylabel('', fontsize=12)
plt.xticks(rotation=45) 
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x="Runtime", y="IMDB Score",data=df, palette='mako')
plt.title('Relation between duration and calification by IMDB Score', fontsize=16)
plt.xlabel('', fontsize=12)
plt.ylabel('', fontsize=12)
plt.xticks(rotation=45) 
plt.show()
```

```python
language_order = df['Language'].value_counts().sort_values(ascending=False).index
top_10_languages = language_order[:10]

plt.figure(figsize=(10,6))
sns.boxplot(x="Language", y="IMDB Score",data=df, palette='mako', order=top_10_languages)
plt.title('Top 10 languages by callification', fontsize=16)
plt.xlabel('', fontsize=12)
plt.ylabel('', fontsize=12)
plt.xticks(rotation=45) 
plt.tight_layout()
plt.show()
```

```python
director_order = df['director'].value_counts().sort_values(ascending=False).index
top_10_director = director_order[:10]

plt.figure(figsize=(10,6))
sns.barplot(x="director", y="IMDB Score",data=df, palette='mako', order=top_10_director)
plt.title('Top 10 directors by calification', fontsize=16)
plt.xlabel('', fontsize=12)
plt.ylabel('', fontsize=12)
plt.xticks(rotation=45) 
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1,2,figsize=(10,3))
sns.countplot(data=df_Netflix[df_Netflix['type'] == 'Movie'], x='rating', ax=axes[0], palette='Blues')
axes[0].set_title('Movies distribution by rating', fontsize=14)
axes[0].set_xlabel('Rating', fontsize=12)
axes[0].set_ylabel('Quantity', fontsize=12)
axes[0].tick_params(axis='x', rotation=45)

sns.countplot(data=df_Netflix[df_Netflix['type'] == 'TV Show'], x='rating', ax=axes[1], palette='Reds')
axes[1].set_title('TV-Shows distribution by rating', fontsize=14)
axes[1].set_xlabel('Rating', fontsize=12)
axes[1].set_ylabel('Quantity', fontsize=12)
axes[1].tick_params(axis='x', rotation=45)
```

```python
fig, axes = plt.subplots(2, 3, figsize = (18, 10))
top_6_countries = df_Netflix['country'].value_counts().nlargest(6).index

axes = axes.flat
for i in range(6):
    sns.countplot(data=df_Netflix[df_Netflix['country']==top_6_countries[i]],x='type',ax=axes[i], palette='Set3')
    axes[i].set_title(f'Movies and TV shows distribution {top_6_countries[i]}', fontsize=14)
    axes[i].set_xlabel('Type', fontsize=12)
    axes[i].set_ylabel('Quantity', fontsize=12)
    axes[i].tick_params(axis='x', rotation=45)  
```

```python
country_order = df['country'].value_counts().sort_values(ascending=False).index
top5 = country_order[:5]

sns.countplot(x="rating",data=df, palette='hls', order=rating_order)
plt.title('Rate distribution', fontsize=16)
plt.xlabel('', fontsize=12)
plt.ylabel('Quantity', fontsize=12)
plt.xticks(rotation=45) 
plt.show()
```

```python
anio_mas_reciente = df_Netflix['release_year'].max()

df_ultimos_10_anios = df_Netflix[df_Netflix['release_year'] >= (anio_mas_reciente - 10)]
top_5_ratings = df_ultimos_10_anios['rating'].value_counts().nlargest(5).index
df_top_5_ratings = df_ultimos_10_anios[df_ultimos_10_anios['rating'].isin(top_5_ratings)]
df_counts = df_top_5_ratings.groupby(['release_year', 'rating']).size().reset_index(name='counts')
plt.figure(figsize=(12,6))
sns.lineplot(
    data=df_counts, 
    x='release_year', 
    y='counts', 
    hue='rating', 
    palette='Set2',
    errorbar= None,
    linewidth = 1,
    linestyle = "dashed", 
    marker = "o"
)

plt.title('Distribución de las 5 Primeras Calificaciones a lo Largo de los Últimos 10 Años', fontsize=16)
plt.xlabel('Año de Lanzamiento')
plt.ylabel('Cantidad de Títulos', fontsize=12)
plt.xticks(rotation=45)

# Mostrar el gráfico
plt.tight_layout()  # Mejora el ajuste del layout
plt.show()
```
